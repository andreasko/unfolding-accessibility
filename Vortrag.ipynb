### 1. Hartmuts SI-Modell

Hartmuts SI-Modell basiert auf der der Gleichung
\begin{equation}
C_T = \prod_{i=1}^T (A_i + 1)
\end{equation}
und wird auf das "social_contacts" und "sexual_contacts" Netzwerk angewendet.
In einem SI-Modell ist naturgemäß die Dichte der gesamten Zugangsmatrix gleich der zeitlich Abhängigen.

## 2. SIR - Modell, Sexualnetzwerk

Es werden zwei Eigenschaften untersucht:
<ul>
<li> a) WSK dass ein beliebiger Knoten zum Zeitpunkt infiziert ist, wenn die Ausbruchsquelle ebenfalls ein beliebiger Knoten sein kann. Diese Angabe kann als Maß für die Ausbruchsgröße einer Epedemie und deren Verlauf genommen werden.
$$\mathcal{P}(n) = \text{nnz}(C(n))/N^2$$

<li> b) Die gleiche Größe $$ P(n) $$ wird für das aggregierte, statische Netzwerk berechnet und mit dem oberen Wert verglichen.
$$c = \mathcal{P}(\infty)/P(\infty) $$
<ul\>

```python
memories = np.arange(100,2300,100)
densities = np.zeros((len(memories),2234))
for ii in range(len(memories)):
    name = "/users/stud/koher/master/python/2014_03_26/results/m="+str(memories[ii])+"_SIR.npz"
    simulation = np.load(name)
    densities[ii,:] = simulation['total_infection_paths']
np.save('sexualContacts_SIR_total_densities',densities)
```

```python
%matplotlib qt
fig = plt.figure()
ax = fig.add_subplot(111)
t = np.arange(SM.runtime)
ax.set_xlabel(r'time (h)',fontsize=16)
ax.set_ylabel(r'$\rho(P_t)$',fontsize=16)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
SM.load('/users/stud/koher/master/python/2014_03_28/results/m=0_aggregate.npz')
ax.plot(range(SM.runtime),SM.current_infection_paths[0:SM.runtime],'r--',label="Aggragate Paths")

plotM = [3,5,8,9,10,12,13,15]
for ii in plotM:
    ax.plot(t,densities[ii,0:-1],label= 'memory = ' + str(memories[ii]))
ax.set_ylim(bottom=0)

```

```python
%matplotlib qt
plt.rc('font', size=13)
c = np.zeros((1,len(memories)))
#densities = np.load('2014_02_27_SIR_total_densities.npy')
for ii in range(len(memories)):
    c[0,ii] = densities[ii,-2]
SM.load('/users/stud/koher/master/python/2014_03_28/results/m=0_aggregate.npz')
density = SM.current_infection_paths[SM.current_infection_paths>0][-1]
c = c/density

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(memories,c[0,:],'b-',linewidth=2)
ax.set_xlabel(r'infectious period (d)',fontsize=16)
ax.set_ylabel(r'$c = \frac{\rho(\mathcal{P}_t)}{\rho(P_t)}$',fontsize=16)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```
