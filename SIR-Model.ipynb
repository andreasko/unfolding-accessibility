# Ein Matrix basierter SIR Formalismus

Zu allererst definieren wir einen Erreichbarkeitsgraphen in der Form, dass eine gerichtete Kante genau dann zwei Knoten verbindet, wenn zum Beobachtungszeitpunkt ein temporärer Pfad zwischen ihnen existiert. Die entsprechende Adjazenzmatrix könnte man der Einfachheit halber Infektions- oder Inzidenzmatrix nennen, weil sie die neuinfizierten Knoten angibt. Diese (boolsche) Matrix lässt sich nun rekursiv berechnen:

\begin{align}
 I(T+1) &= \bigvee_{t=0}^N A_{T+1} \cdot I(T-t) \land \bigwedge_{t=0}^T \overline{I(T-t)}\\
 I(0)   &= \mathbb{1}
\end{align}

### Initialisierung

```python
import numpy as np
from scipy import sparse as sp

# Definiere die Adjazenzmatrizen im csr-sparse Format
A = []
A.append(sp.csr_matrix(([1,1], [[1,0],[0,1]]),shape=(4,4)))
A.append(sp.csr_matrix(([1,1], [[2,1],[1,2]]),shape=(4,4)))
A.append(sp.csr_matrix(([1,1], [[3,2],[2,3]]),shape=(4,4)))
A.append(sp.csr_matrix(([1,1], [[3,0],[0,3]]),shape=(4,4)))

memory=1     #Die Krankheit ueberdauert einen Zeitschritt
size = 4     #Die Dimension des Netzwerks
runtime = 4  #Beobachtungsdauer

#Definiere die Anfangswerte: Alles Nullmatrizen, bis auf I(0), die durch eine Einheitsmatrix gegeben ist
infected    = [sp.csr_matrix((size,size)) for ii in range(memory+1)]
infected[0] = sp.eye(size,size,format='csr')
allInfected = sp.eye(size,size,format='csr')
cumulative  = sp.eye(size,size,format='csr')
```

### Eine Hilfsfunktion zur besseren Darstellung

```python
def myprint(jj):
    if(jj == -1):
        print 'Anfangswerte'
        print
        for ii in range(memory+1):
            print "Infektionsmatrix I(T-" + str(ii) +")"
            print infected[ii].todense()
            print
        print 'Die kumulative Matrix'
        print allInfected.todense()
        print 
        print '================================================'
        print
    else:
        print 'Schritt ' + str(jj+1)
        print
        print "A("+str(jj+1)+")"
        print A[jj].todense()
        print
        for ii in range(memory+1):
            print "Infektionsmatrix I(T-" + str(ii) +")"
            print infected[ii].todense()
            print
        print 'Die Matrix der aktuell erkrankten Knoten'
        print allInfected.todense()
        print 
        print 'Die kumulative Matrix'
        print cumulative.todense()
        print 
        print '================================================'
        print
```

### Der Kern des SIR - Formalismus

```python
myprint(-1)
for jj in range(0,runtime):
    newInfected      = A[jj].dot(allInfected)                         # Infizierte Knoten stecken ihre Nachbarn an
    newInfected.data = np.ones_like(newInfected.data)                 # Stelle sicher, dass die Matrix boolesch ist
    newInfected      = newInfected - newInfected.multiply(cumulative) # Von den neuinfizierten Knoten werden diejenigen entfernt,
                                                                 # die schon in einer beliebigen anderen Klasse enthalten sind.
    allInfected = allInfected - infected[-1]                     # Aus der Klasse aller infizierter Knoten fallen die aeltesten
    allInfected = allInfected + newInfected                      # heraus und die neuen werden dazu addiert.
    cumulative  = cumulative  + newInfected
    
    for ii in range(memory,0,-1):                      # Alle infizierten Knoten altern
        infected[ii] = infected[ii-1]                  # indem sie in die naechst hoehere Klasse verschoben werden
    infected[0] = newInfected
    
    myprint(jj)
```
