# Ein Matrix basierter SIR Formalismus

\begin{align}
I_1(n+1) &= \bigvee_{i=1}^N A_n I_i(n) \land \bigwedge_{i=1}^N \bar{I}_i(n) \land \bar{R}(n)\\
I_2(n+1) &= I_1(n)\\
\vdots\\
R(n+1) &= R(n) \lor I_N(n)
\end{align}

### Initialisierung

```python
import numpy as np
from scipy import sparse as sp

# Definiere die Adjazenzmatrizen im csr-sparse Format
A[0] = sp.csr_matrix(([1,1], [[1,0],[0,1]]),shape=(4,4))
A[1] = sp.csr_matrix(([1,1], [[2,1],[1,2]]),shape=(4,4))
A[2] = sp.csr_matrix(([1,1], [[3,2],[2,3]]),shape=(4,4))
A[3] = sp.csr_matrix(([1,1], [[3,0],[0,3]]),shape=(4,4))

memory=1     #Es gibt nur zwei Infektionsklassen
size = 4     #Die Dimension des Netzwerks
runtime = 4  #Zeitschritte

#Definiere die Anfangswerte: Alles Nullmatrizen, bis auf die erste Infektionsklasse, die durch eine Einheitsmatrix gegeben ist
infclass = [sp.csr_matrix((size,size)) for ii in range(memory+2)]
infclass[0] = sp.eye(size,size,format='csr')
access = sp.eye(size,size,format='csr')
```

### Eine Hilfsfunktion zur besseren Darstellung

```python
def myprint(jj):
    if(jj == 0):
        print 'Anfangswerte'
        print
        for ii in range(memory+1):
            print "infclass " + str(ii)
            print infclass[ii].todense()
            print
        print 'Recovered class'
        print infclass[memory + 1].todense()
        print
        print 'Accessibility Matrix'
        print access.todense()
        print 
        print '================================================'
        print
    else:
        print 'Schritt ' + str(jj)
        print
        print "A("+str(jj)+")"
        print A[jj].todense()
        print
        for ii in range(memory+1):
            print "infclass " + str(ii)
            print infclass[ii].todense()
            print
        print 'Recovered class'
        print infclass[memory + 1].todense()
        print
        print 'Accessibility Matrix'
        print access.todense()
        print 
        print '================================================'
        print
```

### Der Kern des Formalismus

```python
myprint(0)
for jj in range(0,runtime):
    newinf = sp.csr_matrix((size,size))                # Die Neuinfizierten Knoten werden bestimmt:
    for ii in range(0,memory+1):                       # Die infizierten Knoten aller Altersklassen werden durchgegangen 
        newinf = newinf + A[jj].dot(infclass[ii])      # Jeder infektioese Knoten steckt seinen Nachbarn an
    newinf.data[newinf.data>=0] = 1                    # Die Matrizen sollen nur boolsche Werte enthalten
    
    infclass[-1] = infclass[-1] + infclass[-2]         # Die Knoten der letzten Altersklasse gehen zu den "Recovereds" ueber
    for ii in range(memory,0,-1):                      # Alle anderen Klassen altern
        infclass[ii] = infclass[ii-1]                  # indem sie in die naechst hoehere Klasse verschoben werden
    
    newinf = newinf - newinf.multiply(access)          # Von den neuinfizierten Knoten werden diejenigen entfernt,
    infclass[0] = newinf                               # die schon in einer beliebigen anderen Klasse enthalten sind.
    access = access + newinf                           # Schliesslich wird die Zugangsmatrix aktualisiert
    
    myprint(jj)
```
